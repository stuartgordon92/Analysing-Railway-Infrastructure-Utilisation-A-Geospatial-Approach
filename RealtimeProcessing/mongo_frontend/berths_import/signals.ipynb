```python
import geopandas as gpd
from shapely import wkt
import pandas as pd
from ast import literal_eval
```

```python
df = pd.read_csv('signals.csv')
df['geometry'].isna().sum()
```

```python
len(df)
```

```python
df = df.where(pd.notnull(df), None)
df['geometry'] = wkt.loads(df['geometry'])
```

```python
gdf = gpd.GeoDataFrame(df, geometry='geometry',crs='epsg:27700')
```

```python
signals.head(30)
```

```python
signals = gdf[['ASSETID', 'Asset Desc 1', 'Colloquial Name 1', 'Asset Desc 2', 'ELR', 'Track ID', 'geometry', 'Potential Berth', 'Berth']]
signals['Berth'] = signals.apply(lambda row : literal_eval(row['Berth'])[0] if len(literal_eval(row['Berth'])) else None, axis = 1)
signals = signals.fillna(value={"Potential Berth": 0})
signals['Found'] = signals.apply(lambda row : True if row['Potential Berth'] != 0 else False, axis=1)
```

```python
output_path = r'signals_output.geojson'
signals.to_file(output_path, driver="GeoJSON") 
```
